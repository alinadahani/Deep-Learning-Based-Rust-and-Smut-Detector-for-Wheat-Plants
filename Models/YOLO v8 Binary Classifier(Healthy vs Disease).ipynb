```python
from ultralytics import YOLO
from glob import glob

# Load a model
model = YOLO('/content/gdrive/MyDrive/YOLO v8 models/binary classifier(healthy vs disease)/runs/classify/train/weights/best.pt')  # load a custom model
test_images = glob('/content/gdrive/MyDrive/YOLO v8 models//Test data/healthy/1.jpg')

healthy_count = 0
diseased_count = 0

for image_path in test_images:
    # Predict with the model
    results = model(image_path)  # predict on an image

    # Accessing top1 class
    probs = results[0].probs.numpy()


    top1_class = probs.top1

    # Assuming class 0 is diseased and class 1 is healthy
    if top1_class == 0:  # If the top1 class is diseased
        diseased_count += 1
    else:
        healthy_count += 1

print(f'Healthy count: {healthy_count}')
print(f'Diseased count: {diseased_count}')
```
